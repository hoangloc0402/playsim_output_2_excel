```python
import os
from typing import Dict, List, Generator

import pandas as pd
from pprint import pprint

from yaml_utils import (
    load_yaml_file,
    list_all_public_dirs,
    list_yaml_paths_inside_exec_dir,
) 
from pandas_utils import (
    config_2_dataframe,
#     kpi_2_dataframe,
) 
```

```python
ROOT_PATH = '/home/loc/nas_data/playsim'

EXECUTION_DIRS = list_all_public_dirs(ROOT_PATH)
```

```python
exec_dir_2_yaml_paths = dict()
for exec_dir in EXECUTION_DIRS:
    exec_path = os.path.join(ROOT_PATH, exec_dir)
    yaml_paths = list_yaml_paths_inside_exec_dir(exec_path)
    exec_dir_2_yaml_paths[exec_dir] = yaml_paths
```

```python
for exec_dir, yaml_paths in exec_dir_2_yaml_paths.items():
    print('START PROCESSING FOLDER: ', exec_dir)
    
    kpi_data: List[Dict] = list()
    config_data: List[Dict] = list()
        
    for yaml_path in yaml_paths:
        kpi_data.append(
            load_yaml_file(
                os.path.join(ROOT_PATH, exec_dir, yaml_path.kpi)
            ))
                                         
        config_data.append(
            load_yaml_file(
                os.path.join(ROOT_PATH, exec_dir, yaml_path.config)
            ))
        
    break
    print('FINISHED PROCESSING FOLDER: ', exec_dir)
```
